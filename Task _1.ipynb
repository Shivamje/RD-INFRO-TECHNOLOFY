```python
#Load the Dataset
import pandas as pd
#Load dataset from CSV file
df=pd.read_csv('Dataset.csv')
```

#Understand the Data

```python
#See the firt few rows
print(df.head(4))
```

```python
#overview of column type and missing values
print(df.info())
```

#Handle Missing values

```python
#Check the missing value
print(df.isnull().sum())
```

```python
#Remove rows with missing value
#Remove rows with any missing values
data_cleaned=df.dropna()
```

```python
#check again no of missing values
print(data_cleaned.isnull().sum())
```

```python
#Fill Missing values
#Fill missing value in numerical colimns(with median)
df['SeniorCitizen'] = df['SeniorCitizen'].fillna(df['SeniorCitizen'].median())
df['tenure'] = df['tenure'].fillna(df['tenure'].median())
df
```

```python
#Fill missing values in categorical columns (with  mode)
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
    print(df.head(10))
```

```python
#Encode Categorical Variable
from sklearn.preprocessing import LabelEncoder
label_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
le = LabelEncoder()

for col in label_cols:
    df[col] = le.fit_transform(df[col])
```

```python
print(df.head(5))
```

```python
#One-Hot Encode multi-category columns
df_encoded = pd.get_dummies(df.drop(columns=['customerID']), drop_first=True)
```
