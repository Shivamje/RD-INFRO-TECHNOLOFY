```python
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#Load data
df = pd.read_csv("Dataset.csv")
```

```python
df
```

```python
#Encode categorical variables
label_enc = LabelEncoder()
df['Churn'] = label_enc.fit_transform(df['Churn'])  # Yes=1, No=0
df['gender'] = label_enc.fit_transform(df['gender'])
df['Contract'] = label_enc.fit_transform(df['Contract'])
```

```python
df
```

```python
#Box plot - Monthly Charges by Churn
plt.figure(figsize=(8, 5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df,color="darkblue")
plt.title("Monthly Charges by Churn")
plt.xlabel("Churn (0=No, 1=Yes)")
plt.ylabel("Monthly Charges")
plt.show()
```

```python
#Violin plot - Tenure by Contract Type
plt.figure(figsize=(8, 5))
sns.boxplot(x='Contract', y='tenure', data=df,color="red")
plt.title("Tenure Distribution by Contract Type")
plt.xlabel("Contract Type (Encoded)")
plt.ylabel("Tenure")
plt.show()
```

```python
#Pair Plot - Relationships between features
features = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']
df = df[features].dropna()  # Drop rows with missing values in selected features
sns.pairplot(df, hue='Churn')
plt.suptitle("Pair Plot of Key Features Colored by Churn", y=1.02)
plt.show()
```
